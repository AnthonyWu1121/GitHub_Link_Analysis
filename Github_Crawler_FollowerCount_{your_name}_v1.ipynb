```python
def fetch_url_count(url):
    response = requests.get(url, headers=headers, params={'per_page': 1})

    if response.status_code == 200:
        if 'last' in response.links:
            # Get the URL of the last page
            url = response.links['last']['url']
            # Extract the page query parameter from the URL
            total_pages = int(url.split('page=')[-1])
            # Calculate the total count
            url_count = (total_pages - 1) * 1 + len(response.json())
        else:
            # If there's no 'last' link, there's only one page
            url_count = len(response.json())
    else:
        print("Error code:", response.status_code)
        if (response.status_code != 403) and (response.status_code != 429):
          return -1

    return url_count
```

## end ======================

```python
# TODO: Remember the user count in the previous loop and write it here in the next loop.
user_count_start = -1 # initialized as -1
user_count = 0

# TODO: Modify the path to your json file.
repo_path = parent_folder + "/Graph/clean_repo_user_part_{here}.json"

with open(repo_path, 'r') as json_file:
  data = json.load(json_file)

for user in data:
  if user_count <= user_count_start:
    print(f"User: {user_count}")
  else:
    user_data = {
        'username': user,
        'followers_count': fetch_url_count(f"https://api.github.com/users/{user}/followers")
    }
    # TODO: Modify the path to your json file.
    with open(parent_folder + '/crawler code/follower_count/follower_count_{here}.json', 'a') as json_file:
      json.dump(user_data, json_file, indent=4)
      json_file.write('\n')

  print(f"User {user_count} finished.")
  user_count += 1
```
