```python
def fetch_url_count(url):
    response = requests.get(url, headers=headers, params={'per_page': 1})

    if response.status_code == 200:
        if 'last' in response.links:
            # Get the URL of the last page
            url = response.links['last']['url']
            # Extract the page query parameter from the URL
            total_pages = int(url.split('page=')[-1])
            # Calculate the total count
            url_count = (total_pages - 1) * 1 + len(response.json())
        else:
            # If there's no 'last' link, there's only one page
            url_count = len(response.json())
    else:
        print("Error code:", response.status_code)
        if (response.status_code != 403) and (response.status_code != 429):
          return -1

    return url_count
```

```python
import json
import glob

# Load the list of usernames from usernames.json
with open('usernames.json', 'r') as file:
    usernames_list = json.load(file)

# Function to load data from a JSON file
def load_json(filename):
    with open(filename, 'r') as file:
        return json.load(file)

# List all data.json files
data_files = [""]

# Combine all user objects from the data files
combined_user_objects = []
for data_file in data_files:
    combined_user_objects.extend(load_json(data_file))

# Remove duplicates based on the 'username' field
unique_user_objects = {user['username']: user for user in combined_user_objects}.values()

# Extract the unique usernames from the list of user objects
usernames_in_data = {user['username'] for user in unique_user_objects}

# Find missing usernames
missing_usernames = [username for username in usernames_list if username not in usernames_in_data]

# Print the missing usernames
if missing_usernames:
    print("The following usernames are missing:")
    for username in missing_usernames:
        print(username)
else:
    print("All usernames are found in the data.")

# Save the combined and unique user objects to a new file
with open(parent_folder + 'combined_cleaned_data.json', 'w') as file:
    json.dump(list(unique_user_objects), file, indent=4)

print("Combined and cleaned data saved to combined_cleaned_data.json")
```
